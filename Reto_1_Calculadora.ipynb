```python
# 1. Declaración de variables (Tipos de datos: String y Enteros)
mi_nombre = "Kalashnikov"  # Dato tipo String (str)
edad_actual = 28           # Dato tipo Integer (int) - Cambia esto por tu edad
anio_actual = 2026         # Dato tipo Integer (int)

# 2. Operación matemática (Cálculo del año de nacimiento)
anio_nacimiento = anio_actual - edad_actual

# 3. Mostrar el resultado en pantalla usando print()
print("¡Hola a todos!")
print("Mi nombre es", mi_nombre)
print("Tengo", edad_actual, "años.")
print("Si restamos mi edad al año actual, mi año de nacimiento es el:", anio_nacimiento)
```
